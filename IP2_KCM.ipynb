### Question 2 - create two lists, one called x, one called y, each with 5 numbers in them.
- a. Plot them using plt.plot()
- b. Create a plot using plt.subplots() and then add x & y values

```python
xlist=[2,4,5,6,7]
ylist=[6,7,5,9,8]
#a 
plt.plot(xlist, color='indigo')
plt.title("X List")
plt.tick_params(axis="x",
               bottom=False,
               labelbottom=False)
plt.show()
plt.plot(ylist,color='red')
plt.title("Y List")
plt.tick_params(axis="x",
               bottom=False,
               labelbottom=False)
plt.show()
#b
fig=plt.figure(figsize=(10,6))
fig,ax1=plt.subplots()
ax1.plot(xlist,color='purple')
ax1.tick_params(axis="x",
               bottom=False,
               labelbottom=False)
ax1.set_ylabel("X List",
              color="purple",
              size="x-large")
ax2=ax1.twinx()
ax2.plot(ylist,color="pink")
ax2.set_ylabel("Y List",
              color="pink",
              size="x-large")
fig.suptitle("X List vs Y List")
plt.show()
```

### Question 3 - Create an 1D array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X.
- a. Create a plot using plt.subplots() and plot X versus X2
- b. Create a scatter plot of X versus the exponential of X (np.exp(X))

```python
#a
x=np.arange(101)
fig=plt.figure(figsize=(10,6))
fig,ax1=plt.subplots()
ax1.plot(x,color='orange')
ax1.tick_params(axis="x",
               bottom=False,
               labelbottom=False)
ax1.set_ylabel("X List",
              color="orange",
              size="x-large")
ax2=ax1.twinx()
ax2.plot(x**2,color="blue")
ax2.set_ylabel("Y List",
              color="blue",
              size="x-large")
fig.suptitle("X vs X^2")
plt.show()
#b
plt.scatter(x,np.exp(x))
plt.xlabel("X")
plt.ylabel("Exponential of X")
plt.title("Scatter Plot of X vs Exponential of X")
plt.show()
```

### Question 4 - Import car-sales.csv  into a DataFrame called car_sales and investigate about the dataset.

```python
car_sales=pd.read_csv('car-sales.csv')
car_sales.head()
```

### Question 5 - Remove all the symbols from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’)

```python
car_sales['Price']=car_sales['Price'].str.replace('$','')
car_sales['Price']=car_sales['Price'].str.replace(',','')
car_sales['Price']=car_sales['Price'].astype('int64')
car_sales.head()
car_sales.info()
```

### Question 6 - Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

### Question 7 - Add a column called 'Sale Date' which lists a series of successive dates starting from today

```python
car_sales['Sale Date']=pd.date_range('2/21/2024',periods=len(car_sales))
car_sales.head()
```

### Question 8 - Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column

```python
plt.plot(car_sales['Price'],car_sales['Total Sales'])
plt.xlabel("Price")
plt.ylabel("Total Sales")
plt.title("Price vs Total Sales Line Plot")
plt.show()
```

### Question 9 - Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.xlabel("Odometer (KM)")
plt.ylabel('Price')
plt.title("Odometer (KM) vs Price Scatter Plot")
plt.show()
```

### Question 10 - Create a matplotlib bar graph of the 'Make' vs 'Odometer (KM)' columns

```python
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.ylabel("Odometer (KM)")
plt.xlabel('Make')
plt.title("Make vs Odometer (KM) Bar Graph")
plt.show()
```

### Question 11 - Create a histogram of the 'Make' column

```python
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title("Histogram of Car Make")
plt.show()
```

### Question 12 - Create a subplot of multiple plots ( step 8,9,10,11).

```python
fig=plt.figure(figsize=(15,7))
fig.suptitle("Graphs 8-11")
ax1=fig.add_subplot(221)
ax1.set_title("Price vs Total Sales Line Plot")
ax1.plot(car_sales['Price'],car_sales['Total Sales'],color='red')
ax2=fig.add_subplot(222)
ax2.set_title("Odometer (KM) vs Price Scatter Plot")
ax2.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
ax3=fig.add_subplot(223)
ax3.set_title("Make vs Odometer (KM) Bar Graph")
ax3.bar(car_sales['Make'],car_sales['Odometer (KM)'])
ax4=fig.add_subplot(224)
ax4.set_title("Histogram of Car Make")
ax4.hist(car_sales['Make'])
plt.show()
```

### Question 13 - Create a box plot of Colour and Total Sales using seaborn plot

```python
sns.boxplot(x=car_sales['Colour'],y=car_sales['Total Sales'],data=car_sales)
plt.title("Box Plot of Colour and Total Sales")
plt.show()
```

### Question 14 - Find out the relationship between "Total Sales" and "Odometer (KM)" using linear regression with no error bar and car make will determine the hue

```python
sns.lmplot(x='Total Sales',y='Odometer (KM)',hue='Make',ci=None, data=car_sales)
plt.title("Linear regression of Total Sales vs Odometer (KM)")
plt.show()
```
